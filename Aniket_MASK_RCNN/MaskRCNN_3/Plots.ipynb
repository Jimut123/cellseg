```python
import json
import matplotlib.pyplot as plt
with open('history_network_heads.json', 'r') as f:
    history = json.load(f)
print (len(history))
```

```python
for item in history:
    print(item)
```

```python
val_loss = []
for item in history['val_loss']:
    val_loss.append(history['val_loss'][item])

val_rpn_class_loss = []
for item in history['val_rpn_class_loss']:
    val_rpn_class_loss.append(history['val_rpn_class_loss'][item])
    

val_rpn_bbox_loss = []
for item in history['val_rpn_bbox_loss']:
    val_rpn_bbox_loss.append(history['val_rpn_bbox_loss'][item])

val_mrcnn_class_loss = []
for item in history['val_mrcnn_class_loss']:
    val_mrcnn_class_loss.append(history['val_mrcnn_class_loss'][item])

val_mrcnn_bbox_loss = []
for item in history['val_mrcnn_bbox_loss']:
    val_mrcnn_bbox_loss.append(history['val_mrcnn_bbox_loss'][item])

val_mrcnn_mask_loss = []
for item in history['val_mrcnn_mask_loss']:
    val_mrcnn_mask_loss.append(history['val_mrcnn_mask_loss'][item])


loss = []
for item in history['loss']:
    loss.append(history['loss'][item])


rpn_class_loss = []
for item in history['rpn_class_loss']:
    rpn_class_loss.append(history['rpn_class_loss'][item])


rpn_bbox_loss = []
for item in history['rpn_bbox_loss']:
    rpn_bbox_loss.append(history['rpn_bbox_loss'][item])


mrcnn_class_loss = []
for item in history['mrcnn_class_loss']:
    mrcnn_class_loss.append(history['mrcnn_class_loss'][item])


mrcnn_bbox_loss = []
for item in history['mrcnn_bbox_loss']:
    mrcnn_bbox_loss.append(history['mrcnn_bbox_loss'][item])

mrcnn_mask_loss = []
for item in history['mrcnn_mask_loss']:
    mrcnn_mask_loss.append(history['mrcnn_mask_loss'][item])



plt.figure(figsize=(12,6))
plt.title('Loss', fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)

plt.plot(loss,color='blue', linewidth = 3.0)
plt.plot(val_loss,color='orange', linestyle = '-.', dashes=(5, 1), linewidth = 3.0)

plt.plot(rpn_class_loss,color='red', linestyle = '--', dashes=(5, 4), linewidth = 3.0)
plt.plot(val_rpn_class_loss,color='green', linestyle = '--', dashes=(5, 2),  linewidth = 3.0)

plt.plot(rpn_bbox_loss,color='navy', linestyle = '--', dashes=(5, 3), linewidth = 3.0)
plt.plot(val_rpn_bbox_loss,color='skyblue',  linewidth = 3.0)

plt.plot(mrcnn_class_loss,color='deeppink', linestyle = '--', dashes=(5, 5), linewidth = 3.0)
plt.plot(val_mrcnn_class_loss,color='hotpink',  linewidth = 3.0)

plt.plot(mrcnn_bbox_loss,color='purple', linestyle = '--', dashes=(5, 6), linewidth = 3.0)
plt.plot(val_mrcnn_bbox_loss,color='violet',  linewidth = 3.0)

plt.plot(mrcnn_mask_loss,color='sienna', linestyle = '--', dashes=(5, 6), linewidth = 3.0)
plt.plot(val_mrcnn_mask_loss,color='chocolate',  linewidth = 3.0)

# plt.plot(mrcnn_mask_loss,color='dodgerblue', linestyle = '--', dashes=(5, 6), linewidth = 3.0)
# plt.plot(val_mrcnn_mask_loss,color='steelblue',  linewidth = 3.0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#plt.legend(['AttnR2Unet','R2Unet','AttnUnet','MultiResUnet','Mod UNet','UNet', 'Proposed Attn 1'],loc="lower right")
lgd = plt.legend(['Loss','Validation Loss','RPN Class Loss','Validation RPN Class Loss',
                 'RPN Bbox Loss','Validation RPN Bbox Loss', 'MRCNN Class Loss', 'Validation MRCNN Class Loss',
                 'MRCNN Bbox Loss','Validation MRCNN Bbox Loss','MRCNN Mask Loss','Validation MRCNN Mask Loss'],
                 loc="lower right",
          prop={'family':'DejaVu Serif', 'size':20}, bbox_to_anchor=(1.6, -0.1))
plt.savefig('MRCNN_Train_Losses.eps',  bbox_extra_artists=(lgd,), bbox_inches='tight')
#val_rpn_class_loss
#val_mrcnn_class_loss
#rpn_class_loss
#mrcnn_class_lossplt.savefig('MRCNN_Train_Losses.png',  bbox_extra_artists=(lgd,), bbox_inches='tight')
```
