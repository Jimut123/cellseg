```python
import json
import matplotlib.pyplot as plt
with open('history_200e.json', 'r') as f:
    history = json.load(f)
print (len(history))
```

```python
for item in history:
    print(item)
```

```python
accuracy = []
for item in history['accuracy']:
    accuracy.append(history['accuracy'][item])

val_accuracy = []
for item in history['val_accuracy']:
    val_accuracy.append(history['val_accuracy'][item])

plt.figure(figsize=(12,6))
plt.title('Accuracy', fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
plt.plot(accuracy,color='blue', linewidth = 3.0)
plt.plot(val_accuracy,color='orange', linestyle = '-.', dashes=(5, 1), linewidth = 3.0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#plt.legend(['AttnR2Unet','R2Unet','AttnUnet','MultiResUnet','Mod UNet','UNet', 'Proposed Attn 1'],loc="lower right")
lgd = plt.legend(['Accuracy','Validation Accuracy'],loc="lower right",
          prop={'family':'DejaVu Serif', 'size':20}, bbox_to_anchor=(1.44, 0.77))
plt.savefig('Accuracy.eps',  bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python

loss = []
for item in history['loss']:
    loss.append(history['loss'][item])

val_loss = []
for item in history['val_loss']:
    val_loss.append(history['val_loss'][item])

plt.figure(figsize=(12,6))
plt.title('Loss', fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
plt.plot(loss,color='blue', linewidth = 3.0)
plt.plot(val_loss,color='orange', linestyle = '-.', dashes=(5, 1), linewidth = 3.0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#plt.legend(['AttnR2Unet','R2Unet','AttnUnet','MultiResUnet','Mod UNet','UNet', 'Proposed Attn 1'],loc="lower right")
lgd = plt.legend(['Loss','Validation Loss'],loc="lower right",
          prop={'family':'DejaVu Serif', 'size':20}, bbox_to_anchor=(1.37, 0.77))
plt.savefig('Loss.eps',  bbox_extra_artists=(lgd,), bbox_inches='tight')
```
