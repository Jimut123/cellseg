```python
accuracy = []
for item in history['recall']:
    accuracy.append(history['recall'][item])

val_accuracy = []
for item in history['val_recall']:
    val_accuracy.append(history['val_recall'][item])

plt.figure(figsize=(12,6))
plt.title('Recall', fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
plt.plot(accuracy,color='blue', linewidth = 3.0)
plt.plot(val_accuracy,color='orange', linestyle = '-.', dashes=(5, 1), linewidth = 3.0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#plt.legend(['AttnR2Unet','R2Unet','AttnUnet','MultiResUnet','Mod UNet','UNet', 'Proposed Attn 1'],loc="lower right")
lgd = plt.legend(['Recall','Validation Recall'],loc="lower right",
          prop={'family':'DejaVu Serif', 'size':20}, bbox_to_anchor=(1.44, 0.77))
plt.savefig('recall.eps',  bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.savefig('recall.png',  bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
accuracy = []
for item in history['precision']:
    accuracy.append(history['precision'][item])

val_accuracy = []
for item in history['val_precision']:
    val_accuracy.append(history['val_precision'][item])

plt.figure(figsize=(12,6))
plt.title('Precision', fontsize=20, fontname = 'DejaVu Serif', fontweight = 500)
plt.plot(accuracy,color='blue', linewidth = 3.0)
plt.plot(val_accuracy,color='orange', linestyle = '-.', dashes=(5, 1), linewidth = 3.0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#plt.legend(['AttnR2Unet','R2Unet','AttnUnet','MultiResUnet','Mod UNet','UNet', 'Proposed Attn 1'],loc="lower right")
lgd = plt.legend(['precision','Validation precision'],loc="lower right",
          prop={'family':'DejaVu Serif', 'size':20}, bbox_to_anchor=(1.44, 0.77))
plt.savefig('precision.eps',  bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.savefig('precision.png',  bbox_extra_artists=(lgd,), bbox_inches='tight')
```
