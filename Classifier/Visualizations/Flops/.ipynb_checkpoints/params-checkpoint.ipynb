```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
plt.rcParams['text.usetex'] = True
# print(plt.rcParams.keys())

plt.rcParams['text.latex.preamble'] = r"\usepackage{bm} \usepackage{amsmath}"

import seaborn as sns
sns.set_style("darkgrid")

y = [14.78, 21.12, 21.75, 22.06, 42.92, 54.53, 85.43]
x = [1,2,3,4,5,6,7]
labels = [r'$\textbf{VGG 16}$',r'$\textbf{Xception}$', \
          r'$\textbf{ViT}$', r'$\textbf{InceptionV3}$', r'$\textbf{Resnet101}$',\
          r'$\textbf{InceptionResNetV2}$', r'$\textbf{NASNetLarge}$']
width = 0.5
plt.title(r'$\textbf{No. of parameters (in Million)}$',fontsize=20, fontweight = 500)
plt.xticks(x, labels, rotation=90, fontsize=15, fontweight = 500)
plt.bar(x, y, width, color='aquamarine')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig('params.eps',bbox_inches='tight')
plt.savefig('params.png',bbox_inches='tight')
plt.show()
```
