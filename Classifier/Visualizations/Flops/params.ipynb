```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

y = [1.203649, 7.238228, 11.694817, 14.830177, 21.556472, 31.379075, 31.902759, 95.986179, 96.509863]
x = [1,2,3,4,5,6,7,8,9]
labels = ['A','B','C', 'D', 'E', 'F','G','H','I']
width = 0.5
plt.title('No. of parameters in millions ',fontsize=20,fontname = 'DejaVu Serif', fontweight = 500)
plt.xticks(x, labels, rotation=90, fontsize=15, fontname = 'DejaVu Serif', fontweight = 500)
plt.bar(x, y, width, color='aquamarine')
plt.xticks(fontsize=15, fontname = 'DejaVu Serif')
plt.yticks(fontsize=15, fontname = 'DejaVu Serif')
plt.savefig('params.eps',bbox_inches='tight')
plt.savefig('params.png',bbox_inches='tight')
plt.show()
```
