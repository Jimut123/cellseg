```python
val_loss = []
for item in history['val_loss']:
    val_loss.append(history['val_loss'][item])

for item in history_new['val_loss']:
    val_loss.append(history_new['val_loss'][item])


val_rpn_class_loss = []
for item in history['val_rpn_class_loss']:
    val_rpn_class_loss.append(history['val_rpn_class_loss'][item])
    
for item in history_new['val_rpn_class_loss']:
    val_rpn_class_loss.append(history_new['val_rpn_class_loss'][item])


val_rpn_bbox_loss = []
for item in history['val_rpn_bbox_loss']:
    val_rpn_bbox_loss.append(history['val_rpn_bbox_loss'][item])
    
for item in history_new['val_rpn_bbox_loss']:
    val_rpn_bbox_loss.append(history_new['val_rpn_bbox_loss'][item])

    
val_mrcnn_class_loss = []
for item in history['val_mrcnn_class_loss']:
    val_mrcnn_class_loss.append(history['val_mrcnn_class_loss'][item])

for item in history_new['val_mrcnn_class_loss']:
    val_mrcnn_class_loss.append(history_new['val_mrcnn_class_loss'][item])


val_mrcnn_bbox_loss = []
for item in history['val_mrcnn_bbox_loss']:
    val_mrcnn_bbox_loss.append(history['val_mrcnn_bbox_loss'][item])

for item in history_new['val_mrcnn_bbox_loss']:
    val_mrcnn_bbox_loss.append(history_new['val_mrcnn_bbox_loss'][item])


val_mrcnn_mask_loss = []
for item in history['val_mrcnn_mask_loss']:
    val_mrcnn_mask_loss.append(history['val_mrcnn_mask_loss'][item])

for item in history_new['val_mrcnn_mask_loss']:
    val_mrcnn_mask_loss.append(history_new['val_mrcnn_mask_loss'][item])

    
loss = []
for item in history['loss']:
    loss.append(history['loss'][item])

for item in history_new['loss']:
    loss.append(history_new['loss'][item])


rpn_class_loss = []
for item in history['rpn_class_loss']:
    rpn_class_loss.append(history['rpn_class_loss'][item])

for item in history_new['rpn_class_loss']:
    rpn_class_loss.append(history_new['rpn_class_loss'][item])


rpn_bbox_loss = []
for item in history['rpn_bbox_loss']:
    rpn_bbox_loss.append(history['rpn_bbox_loss'][item])

for item in history_new['rpn_bbox_loss']:
    rpn_bbox_loss.append(history_new['rpn_bbox_loss'][item])


mrcnn_class_loss = []
for item in history['mrcnn_class_loss']:
    mrcnn_class_loss.append(history['mrcnn_class_loss'][item])

for item in history_new['mrcnn_class_loss']:
    mrcnn_class_loss.append(history_new['mrcnn_class_loss'][item])

    
mrcnn_bbox_loss = []
for item in history['mrcnn_bbox_loss']:
    mrcnn_bbox_loss.append(history['mrcnn_bbox_loss'][item])

for item in history_new['mrcnn_bbox_loss']:
    mrcnn_bbox_loss.append(history_new['mrcnn_bbox_loss'][item])

    
mrcnn_mask_loss = []
for item in history['mrcnn_mask_loss']:
    mrcnn_mask_loss.append(history['mrcnn_mask_loss'][item])

for item in history_new['mrcnn_mask_loss']:
    mrcnn_mask_loss.append(history_new['mrcnn_mask_loss'][item])
```
