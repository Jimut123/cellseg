```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# all_files = glob.glob('Smear_Slides/*')
all_files = glob.glob('PBC_dataset_normal_DIB_cropped/*')
```

```python
for img_name in all_files:
    img = cv2.imread(img_name)
    black_pixels = np.where(
        (np.logical_and(img[:, :, 0] >= 0, img[:, :, 0] <= 20)) & 
        (np.logical_and(img[:, :, 1] >= 0, img[:, :, 1] <= 20)) & 
        (np.logical_and(img[:, :, 2] >= 0, img[:, :, 2] <= 20))
        )
    # set those pixels to white
    img[black_pixels] = [255, 255, 255]
    
    plt.imshow(img)
    plt.show()
    cv2.imwrite(img_name, img)
```
