```python
import json
import matplotlib.pyplot as plt
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
with open('history_Smear_10_cropped_NASNetLarge_fine_tuned_100e.json', 'r') as f:
    NASNetLarge = json.load(f)
print (len(NASNetLarge))
```

```python
with open('history_Smear_10_cropped_InceptionResNetV2_fine_tuned_100e.json', 'r') as f:
    InceptionResNetV2 = json.load(f)
print (len(InceptionResNetV2))
```

```python
with open('history_Smear_10_cropped_inception_v3_fine_tuned_100e.json', 'r') as f:
    inception_v3 = json.load(f)
print (len(inception_v3))
```

```python
with open('history_Smear_10_cropped_resnet101_fine_tuned_100e.json', 'r') as f:
    resnet101 = json.load(f)
print (len(resnet101))
```

```python
with open('history_Smear_10_cropped_vgg_16_100e.json', 'r') as f:
    vgg_16 = json.load(f)
print (len(vgg_16))
```

```python
loss_NASNetLarge = []
for item in NASNetLarge['loss']:
    loss_NASNetLarge.append(NASNetLarge['loss'][item])


loss_InceptionResNetV2 = []
for item in InceptionResNetV2['loss']:
    loss_InceptionResNetV2.append(InceptionResNetV2['loss'][item])  


loss_inception_v3 = []
for item in inception_v3['loss']:
    loss_inception_v3.append(inception_v3['loss'][item])



loss_resnet101 = []
for item in resnet101['loss']:
    loss_resnet101.append(resnet101['loss'][item])


loss_xception = []
for item in xception['loss']:
    loss_xception.append(xception['loss'][item])

    
loss_vgg_16 = []
for item in vgg_16['loss']:
    loss_vgg_16.append(vgg_16['loss'][item])


    
plt.figure(figsize=(12,6))
plt.title('Training Loss for Smear Slide Cropped Dataset', fontsize=30, fontname = 'DejaVu Serif', fontweight = 500)
plt.plot(loss_NASNetLarge, color='red', linestyle='-.', dashes=(5, 5), linewidth=2.0)
plt.plot(loss_InceptionResNetV2, color='green',  linestyle='--', dashes=(5, 1),  linewidth=2.0)
plt.plot(loss_inception_v3, color='darkorange', linestyle='--', dashes=(5, 1),  linewidth=2.0)
plt.plot(loss_resnet101, color='brown',  linewidth=2.0)
plt.plot(loss_xception, color='deepskyblue', linestyle='-', linewidth=2.0)
plt.plot(loss_vgg_16, color='navy', linestyle=':',  linewidth=2.0)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

lgd = plt.legend(['NASNetLarge','InceptionResNetV2','Inception-v3','Resnet101','Xception','VGG 16'], 
           loc="lower right", prop={'family':'DejaVu Serif', 'size':20}, bbox_to_anchor=(1.43, 0.415))
plt.savefig('Smear_10_cropped_fine_tuned_losses.eps', bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.savefig('Smear_10_cropped_fine_tuned_losses.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
